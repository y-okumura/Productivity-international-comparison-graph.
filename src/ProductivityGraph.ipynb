```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
book = pd.ExcelFile('tabula-intl_comparison_2017_data.xlsx')
```

```python
par_person, par_hour = [book.parse(sheet_name) for sheet_name in book.sheet_names]
```

```python
par_person
```

```python
def parse(sheet):
    data = {int(year[0:-1]):{} for year in sheet.columns if year[-1] == "年"}
    for row in sheet.values:
        for (year, country, productivity) in zip(data.keys(), row[0::2], row[1::2]):
            data[year][country] = productivity

    return pd.DataFrame.from_dict(data)
```

```python
dfs = {"一人当たり": parse(par_person), "時間当たり": parse(par_hour)}
```

```python
dfs["一人当たり"]
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'Noto Sans CJK JP'

for name in dfs.keys():
    dfs[name].to_csv(name + '.tsv', sep="\t", line_terminator="\r\n")
    dfs[name].T.plot()
```
